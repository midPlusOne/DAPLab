```python
#importing the dataset
import pandas as pd
import io
import numpy as np

data = pd.read_csv('/home/mcacnj/Mubassir/Dataset/file1.csv')
print('importing the dataset')
data.head(10)
```

```python
#Missing data
print(data.shape)
df_dropped = data.dropna()
df_dropped.head(10)
```

```python
#Data frame manipulation using numpy
import pandas as pd
import numpy as np

#creating numpy array
array = np.array([[1,1,1],[2,3,8],[3,5,27]])

#creating a list of index names
index_values = ['Row1','Row2','Row3']

#creating a list of colums names
column_values = ['Col1','Col2','Col3']

#creating the dataframe
df = pd.DataFrame(data = array, index = index_values, columns = column_values)

#displaying the dataframes
print(df)

#calculating the trace of a matrix
trace = np.trace(df)
print("\nTrace of given 3x3 matrix : ")
print(trace)
```
