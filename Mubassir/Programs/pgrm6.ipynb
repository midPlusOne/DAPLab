```python
#1.line plot
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2,4,6,8])
y = x * 2
plt.plot(x,y)
plt.show()
```

```python
#2.Box plot
import pandas as pd
import numpy as np
df  = pd.DataFrame(np.random.rand(20,5),columns=['A','B','C','D','E'])
df.plot.box(grid="True")
```

```python
#3.Scatter plot
from matplotlib import pyplot as plt
#x-axis values
x = [74,88,52,72,78]
#y-axis values
y = [23,22,54,64,80]
#function to plot scatter
plt.scatter(x,y)
plt.show()
```

```python
#4.Histogram
from matplotlib import pyplot as plt
#y-axis values
y = [1,15,19,40,65]
plt.hist(y)
plt.show()
```

```python
#5.Heat maps
from pandas import DataFrame
import matplotlib.pyplot as plt
data = [{8,9,3,4},{4,6,9,1},{7,4,1,3},{9,5,1,3},{7,1,3,9}]
Index = ['I1','I2','I3','I4','I5']
Cols = ['C1','C2','C3','C4']
df = DataFrame(data, index = Index, columns = Cols)
plt.pcolor(df)
plt.show()
```
