```python
#Implement a python program to demonstarte 
#a) Importing dataset, 
#b) Cleansing the data, 
#c) Data frame manipulation using numpy
#Importing data set

import pandas as pandas
import numpy as numpy
import io

data = pandas.read_csv('Dataset/dataset.csv')
data.head(10)
```

```python
#sum() method is used to calculate the count
#isnull() method is used to check for null values
data.isnull().sum()
```

```python
#cleansing the dataset
data.drop_duplicates(subset = ['name'],keep = "last", inplace = True)
data.head(10)
```

```python
#missing data
print(data.shape)
#data.shape is used to to print number of columns and rows in the dataset
df_dropped = data.dropna()
df_dropped.head(10)
```

```python
#filling values
data['Hobbies'] = data['Hobbies'].fillna('Soccer')
mean_val = data['Mobile Number'].mean()
data['Mobile Number'] = data['Mobile Number'].fillna(mean_val)
data.head(12)
```

```python
#data frame manipulation using numpy
import pandas as pd
import numpy as np

arr = np.array([[1,1,1],[2,3,8],[3,5,2]])

idx = ['Row 1','Row 2','Row 3']
col = ['Col 1','Col 2','Col 3']

df = pd.DataFrame(data = arr, index = idx, columns = col)

print(df)

trace = np.trace(df)

print("\nTrace of a given 3X3 matrix is: ",trace)
```
